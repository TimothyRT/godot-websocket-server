Tes penggunaan filter Madgwick

```python
import numpy as np
import time
import quaternion
from ahrs.filters import Madgwick
```

```python
def do_something():
    gyro_data = np.array([3.0, -13.0, 3.2])
    accel_data = np.array([4.2, 5.0, -4.0])
    mag_data = np.array([0.3, -4.0, 2.0])
    madgwick = Madgwick(frequency=17.0, gain=0.041, q0=quaternion.as_float_array(quaternion.one))
    new_q = madgwick.updateMARG(madgwick.q0, gyro_data, accel_data, mag_data, 1/17.0)
    print(new_q)
```

```python
t1 = time.time()
do_something()
t2 = time.time()
```

```python
print(t2 - t1)
```

```python
%%timeit
do_something()
```
